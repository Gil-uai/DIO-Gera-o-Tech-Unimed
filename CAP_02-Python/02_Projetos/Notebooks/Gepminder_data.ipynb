# Analise Dados GEPminder ⛳


O base de dados Gapminder contem informações continentes, países, numero de habitantes, idade, epectativa de vida, PIB. Para nosso projeto iremos utilizar esta base de dados para conhecer e praticar alguns comandos utilizando a biblioteca pandas

```python
#importando a biblioteca Pandas
import pandas as pan

```

```python
base = pan.read_csv("/content/Gapminder.csv",error_bad_lines=False, sep=";") #error_bad_lines foi utilizado para ignorar erro existente de coluna fazia / sep inserido para alterar marcado de separação
```

```python
#head é utlizado para exibir uma quantidade de linhas da base de dados
base.head (2)
```

```python
#Comando para alterar o nome das colunas
base = base.rename (columns={"contry":"Pais", "continent":"Continente","year":"Ano","lifeExp":"Expectativa de vida","pop":"População total","gdpPercap":"PIB"})
```

```python
base.head () #Verificação da alteração do nomes das colunas
```

```python
#verificação da quantidade total de linhas e colunas
base.shape
```

```python
# Verificação do nome da colunas da base de dados
base.columns
```

```python
# Exibir o tipo de dados de cada coluna
base.dtypes
```

```python
# Retorna X ultimas linhas da base
base.tail(10)
```

```python
# Exibir as estatisticas gerais da base de dados
base.describe ()
```

```python
# Exibir os valores unicos de cada coluna
base["Continente"].unique()
```

```python
# Realiza um filtro na coluna Continente
Oceania = base.loc[base["Continente"] == "Oceania"]
Oceania.head(3)
```

```python
Oceania["Continente"].unique()
```

```python
#Realiza o agrupamento dos dados
base.groupby("Continente")["country"].nunique()
```

```python
#Agrupa os dados por ano e informa a media
base.groupby("Ano")["Expectativa de vida"].mean()
```

```python
# Exibir a média do PIB
base["PIB"].mean()
```

```python
# Exibir a soma do PIB
base["PIB"].sum()
```
