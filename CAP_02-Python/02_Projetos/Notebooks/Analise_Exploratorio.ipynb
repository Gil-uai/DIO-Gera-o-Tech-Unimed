```python
#Importando as bibliotecas que ser칚o utilizadas no projeto
import pandas as pand
import matplotlib.pyplot as plotar
plotar.style.use ("dark_background") # definindo o stylo dos graficos que iremos utilizar ao longo da an치lise
```

```python
# Realizando o carregamento da base de dados
from google.colab import files
arquivo = files.upload()
```

```python
# Criando dataframe
base = pand.read_excel("AdventureWorks.xlsx")
```

```python
# Verifica칞칚o rapida do carregamento dos dados
base.head (3)
```

```python
# Verificando o tamanho a planilha (linhas, colunas)
base.shape
```

```python
# Verificando o tipo de dado de cada coluna
base.dtypes
```

```python
# Analisar qual a receita total
base["Valor Venda"].sum() # Tomamos como premissa que a coluna #Valor Venda" 칠 o valor unitario multiplicado pela quantidade
```

```python
# Analisar a quantidade total de itens vendidos
base["Quantidade"].sum()
```

```python
base["Quantidade"].count()
```

```python
# Encontrando a m칠dia de itens vendidos por venda de forma indireta
base["Quantidade"].sum() / base["Quantidade"].count()
```

```python
# Econtrando a m칠dia de itens vendidos por venda atraves de metodo mean
base["Quantidade"].mean()
```

```python
# Analisar qual o custo de venda total
base["Custo"] = base["Custo Unit치rio"].mul(base["Quantidade"]) # criando a coluna Custo de cada venda multiplicando o total vendido pelo custo unitario
```

```python
base.head()
```

```python
base.sample ()
```

```python
#Verificando custo total
base["Custo"].sum()
```

```python
# Criando uma coluna com o lucro por venda
base["Lucro"] = base["Valor Venda"] - base["Custo"]
```

```python
base.sample (3)
```

```python
# Verificando lucro total
base["Lucro"].sum()
```

```python
# Criando coluna com numero de dias para envio do produto
base["Tempo Envio"] = (base["Data Envio"] - base["Data Venda"])
```

```python
base.sample (3)
```

```python
# Alterar a coluna Tempo Envio eliminando a String "days"
# Ao criar a coluna "Tempo Envio" foi adicionada string "days" no final com isso n칚o conseguimos fazer calculo com a mesma
base["Tempo Envio"] = (base["Data Envio"] - base["Data Venda"]).dt.days # agora acrescentamos para considerar somente o dia
```

```python
base.head(1)
```

```python
# Conferindo o tipo de dado da coluna "Tempo Envio"
base["Tempo Envio"].dtypes
```

```python
# Verificando a m칠dia do tempo de envio por marca
base.groupby("Marca")["Tempo Envio"].mean()
```

### Verificar exist칡ncia de Valores nulos

```python
# Vrrificando quantidade de itens faltantes por coluna
base.isnull().sum()
```

### Agrupar Valores

```python
# Verificando o lucro por ano e por marca
base.groupby([base["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
# Alterando a forma de exibi칞칚o
pand.options.display.float_format = '{:20,.2f}'.format # formata os numero float para 2 casas decimais
```

```python
# Resentando o index e transformando os retornos em tabela
lucro_ano = base.groupby([base["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
```

```python
lucro_ano
```

```python
# Verificando a quantidade total vendida por produto
base.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False) # acrescentado sort para colocar em ordem crecente
```

### Realizar an치lises com gr치ficos

```python
# Exibir grafico com total de produtos vendidos por marca
base.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title = "Vendas por Produto");
```

```python
# Exibir grafico de lucro por ano
base.groupby(base["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title = "Lucro x ano")
plotar.xlabel("Ano")
plotar.ylabel("Lucro");
```

```python
# Analisando somente as vendas de 2009
# agrupar somente as vendas de 2009
base_2009 = base[base["Data Venda"].dt.year == 2009]
```

```python
base_2009.head(1)
```

```python
# Verificando na nova base temos somente dados de 2009
base_2009.sample (5)
```

```python
# Exigir grafico do lucro por m칡s no ano de 2009
base_2009.groupby(base_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title = "Lucro x M칡s");
plotar.xlabel ("M칡s")
plotar.ylabel ("Lucro");
```

```python
# Exibir grafico de lucro por marca no ano de 2009
base_2009.groupby(base_2009["Marca"])["Lucro"].sum().plot.bar(title = "Lucro x Marca")
plotar.xlabel ("Marca")
plotar.ylabel ("Lucro");
plotar.xticks(rotation='horizontal');
```

```python
# Exibir grafico de lucro por classe no ano de 2009
base_2009.groupby(base_2009["Classe"])["Lucro"].sum().plot.bar(title = "Lucro x Marca")
plotar.xlabel ("Classe")
plotar.ylabel ("Lucro");
plotar.xticks(rotation='horizontal');
```

```python
# Exibir estatisticas gerais do tempo de envio
base["Tempo Envio"].describe()
```

```python
# Exibir grafico tipo boxplot para o tempo de envio
plotar.style.use ("classic") # Alterado Stilo para melhorar visualiza칞칚o do boxplot
plotar.boxplot(base["Tempo Envio"]);
```
